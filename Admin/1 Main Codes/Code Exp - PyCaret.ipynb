```python
#CodeSection2
import pandas as pd
import numpy as np
from pycaret.classification import *
```

```python
from sklearn.metrics import accuracy_score
```

## 2/ Import Data

```python
#CodeSection3
train = pd.read_csv('/mntDrive/MyDrive/Project MasterMind/1 - Planning Stage/2- Research/6- New Teaching Method/Loan Prediction (Webinar Flow)/Solution Codes/input/train.csv')
test = pd.read_csv('/mntDrive/MyDrive/Project MasterMind/1 - Planning Stage/2- Research/6- New Teaching Method/Loan Prediction (Webinar Flow)/Solution Codes/input/test.csv')
```

## 3/ Create 2 New Features


```python
#CodeSection4
all_data = [train, test]
```

### 3.1 Create First Feature

```python
#CodeSection5
# Total Income is Sum of Applicant Income and Co-applicant Income
for dataset in all_data:
    dataset['Total_Income'] = dataset['ApplicantIncome'] + dataset['CoapplicantIncome']
```

### 3.2 Create Second Feature

```python
#CodeSection6
# Loan by Income is Loan Amount divided by Total Income
for dataset in all_data:
    dataset['Loan_by_Income'] = dataset['LoanAmount']/dataset['Total_Income']
```

## Setup PyCaret

```python
exp_name = setup(data = train,  target = 'Loan_Status', 
                 ignore_features= ['Loan_ID'], 
                 numeric_imputation='median', 
                 normalize= True, 
                 transformation=True, 
                 log_experiment = True, 
                 experiment_name = "Test Exp 1",
                 polynomial_features = True,
                 polynomial_degree = 2,
                 trigonometry_features = True, 
                 feature_interaction = True,
                  fold = 5, 
                  )
```

```python
top3 = compare_models(n_select = 3)
```

```python
rf = create_model('rf', round = 2)
```

```python
tuned_rf = tune_model(rf, optimize = 'Accuracy')
```

```python
plot_model(tuned_rf, plot = 'parameter')
```

## 6/ Check Accuracy of Model on Train Data

```python
pred_holdout = predict_model(tuned_rf)
```

```python
pred_train = predict_model(tuned_rf, data = train)
```

```python
#CodeSection21
# Predict on Train Data
y_train_pred = pred_train['Label']
```

### We can use a Accuracy Function from Metrics
- Check Train Accuracy

```python
#CodeSection23
# Print Train Accuracy
print(f" Train Accuracy : {accuracy_score(train['Loan_Status'], y_train_pred):0.1%}")
```

## 7/ Predict and Submission

### Predict on "Test Data"

```python
pred_test = predict_model(tuned_rf, data = test)
y_test_pred = pred_test['Label']
```

```python
X_test_prep = pred_test['Label']
```

### Create Submission File

```python
#CodeSection25
submission = pd.DataFrame({
    'Loan_ID' : test['Loan_ID'],
    'Loan_Status' : X_test_prep
})
```
